```python
import pandas as pd

# Assuming your 2019 data is in df_2019
attributes = [
     'Social support',
       'Healthy life expectancy', 'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption'
]

# Apply Min-Max scaling to all five columns
for col in attributes:
    min_val = merged_df[col].min()
    max_val = merged_df[col].max()
    
    # Create a new column with the 0-1 scaled score
    merged_df[col + '_0_1_Scale'] = (merged_df[col] - min_val) / (max_val - min_val)

print("Attributes successfully scaled to 0 to 1.")
```

```python
WEIGHT = 20  # 20% of 100

# Create a list of the new 0-1 scaled column names
scaled_cols = [col + '_0_1_Scale' for col in attributes]

# 1. Apply the 20% weight to each scaled attribute
for col in scaled_cols:
    merged_df[col + '_Weighted'] = merged_df[col] * WEIGHT

# 2. Sum the five weighted columns to create the final 100-point index
weighted_cols = [col + '_Weighted' for col in scaled_cols]
merged_df['Custom_Happiness_Index_100'] = merged_df[weighted_cols].sum(axis=1)

print("Custom Index created and summed to a maximum of 100.")
```

```python
# Create a ranking column (rank 1 is the highest score) step 3 
merged_df['Index_Rank'] = merged_df['Custom_Happiness_Index_100'].rank(ascending=False)

# Display the top 10 ranked countries, showing their total score and HLE contribution
final_ranking = merged_df.sort_values(by='Index_Rank').head(10)

print("\n--- Top 10 Countries by Custom 100-Point Index ---")
print(final_ranking[[
    'country', 
    'Index_Rank', 
    'Custom_Happiness_Index_100',
    'Healthy life expectancy_0_1_Scale_Weighted' # HLE Contribution (Ratio)
]])
```

```python
# --- STEP 1: DEFINE THE UNDEFINED VARIABLE ---
# This list MUST contain the exact names of the original columns 
# that you scaled in the previous step.
cols = [
    'Social support',
    'Healthy life expectancy',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
]
# ----------------------------------------------


WEIGHT = 20  # 20% of 100

# Create a list of the new 0-1 scaled column names
# EXAMPLE: 'Social support' becomes 'Social support_0_1_Scale'
scaled_cols = [col + '_0_1_Scale' for col in cols]

# 1. Apply the 20% weight to each scaled attribute
for col in scaled_cols:
    # Use the correct DataFrame name, which should be 'df_2019' from the previous context, 
    # but I will use 'merged_df' as per your provided code for consistency.
    merged_df[col + '_Weighted'] = merged_df[col] * WEIGHT

# 2. Sum the five weighted columns to create the final 100-point index
weighted_cols = [col + '_Weighted' for col in scaled_cols]
merged_df['Custom_Happiness_Index_100'] = merged_df[weighted_cols].sum(axis=1)

print("Custom Index created and summed to a maximum of 100.")
```

```python
import pandas as pd

# 1. Define the columns you want to include in the new DataFrame
# Assuming 'Country or region' is the name of your country column
summary_cols = [
    'country', 
    'Custom_Happiness_Index_100',
    'Index_Rank'
]

# 2. Create the new DataFrame by selecting the desired columns
ranking_df = merged_df[summary_cols].copy()

# 3. Sort the new DataFrame by the Index_Rank (Ascending: Rank 1 is the best)
ranking_df_sorted = ranking_df.sort_values(by='Index_Rank', ascending=True)

# 4. Display the top few rows of the final ranking
print("--- Final Custom Happiness Index Ranking (Top 10) ---")
print(ranking_df_sorted.head(10))
```

```python
# Define the five factor names (must match the original columns used for scaling)
attributes = [
    'Social support',
    'Healthy life expectancy',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
]

# The contribution factor (20% of the total score)
CONTRIBUTION_FACTOR = 0.20

# Create a list to store the names of the new contribution columns
contribution_cols = []

# Loop through each attribute and calculate its 20% average contribution
for attr in attributes:
    new_col_name = f'{attr}_Avg_Contribution'
    
    # Calculate 20% of the total index score for each country
    ranking_df_sorted[new_col_name] = ranking_df_sorted['Custom_Happiness_Index_100'] * CONTRIBUTION_FACTOR
    
    contribution_cols.append(new_col_name)

# ----------------------------------------------------------------------
# 2. Display the Final Ranking with the Full Breakdown
# ----------------------------------------------------------------------

# Define the columns for the final display
display_cols = ['country', 'Index_Rank', 'Custom_Happiness_Index_100'] + contribution_cols

#print("\n--- Final Ranking with Full Theoretical 20% Breakdown (Top 5) ---")
#print(ranking_df_sorted[display_cols].head(10))

# Sort the DataFrame by 'Custom_Happiness_Index_100' in descending order
ranking_df_sorted = ranking_df_sorted.sort_values(
    by='Custom_Happiness_Index_100', 
    ascending=False
)

# Display the entire sorted DataFrame
display(ranking_df_sorted)
```

```python
# 1. Define the list of target countries
target_countries = [
    'Finland', 'Denmark', 'Norway', 'Iceland', 'Netherlands', 
    'Switzerland', 'Sweden', 'New Zealand', 'Canada', 'Austria'
]

# 2. Filter the DataFrame using the .isin() method
filtered_df = ranking_df_sorted[
    ranking_df_sorted['country'].isin(target_countries)
].copy()

# 3. Display the new filtered DataFrame
print("\n--- Filtered Ranking (Top 10 Countries by Name) ---")
display(filtered_df)
```
