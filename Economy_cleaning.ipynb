```python
import pandas as pd
df = pd.read_excel("GDP_dataset.xlsx", sheet_name="Data")
df
```

```python
df.info()
```

```python
df.shape
```

```python
df_2019 = df[df["year"] == 2019]
df_2019
```

```python
economy = df_2019[["country", "pop", "rgdpna", "avh", "hc", "ccon"]]
economy
```

```python
economy = economy.reset_index(drop=True)
economy
```

```python
#Real GDP at constant 2021 national prices = the size of a country’s economy measured in inflation-adjusted 2021 US dollars, making it possible to compare economic performance across years without price changes interfering.

#Real consumption of households and government, at current PPPs (in mil. 2021US$) = This variable measures how much people + the government actually spend, adjusted so you can compare it fairly across countries.

#The Human Capital Index (hc) measures: How much the average worker’s education increases their productivity, based on: years of schooling return on education (how much extra income education brings). The value is usually between 1.0 and 4.0.
```

```python
economy = economy.rename(columns={
    "pop": "population(mil.)",
    "rgdpna": "GDP(mil./$)",
    "avh": "av_Working_h",
    "hc": "human_capital_index",
    "ccon": "real_consumption(mil./$)"
})
economy
```

```python
columns = ["population(mil.)", "av_Working_h", "human_capital_index", "real_consumption(mil./$)"]
economy["population(mil.)"] = economy["population(mil.)"].round(3)
economy["GDP(mil./$)"] = economy["GDP(mil./$)"].astype(float)
economy["GDP(mil./$)"] = economy["GDP(mil./$)"].round(2)
economy["av_Working_h"] = economy["av_Working_h"].round(0)
economy["human_capital_index"] = economy["human_capital_index"].round(2)
economy["real_consumption(mil./$)"] = economy["real_consumption(mil./$)"].round(1)
economy
```

```python
#Cleaning
```

```python
economy.isnull().sum()
```

```python
#Filling missing values with mean
economy["av_Working_h"] = economy["av_Working_h"].fillna(
    economy["av_Working_h"].mean()
)

economy["human_capital_index"] = economy["human_capital_index"].fillna(
    economy["human_capital_index"].mean()
)
economy
```

```python
economy.dtypes
```

```python
economy["av_Working_h"] = economy["av_Working_h"].astype(int)
economy["human_capital_index"] = economy["human_capital_index"].round(2)
economy
```

```python
#Check for outliers
economy[economy["population(mil.)"] <= 0]
```

```python
economy.dtypes
```

```python
economy = economy.copy()
# GDP per capita
economy["GDP_per_capita"] = (
    economy["GDP(mil./$)"] / economy["population(mil.)"]
)

# Consumption per capita
economy["real_consumption_per_capita"] = (
    economy["real_consumption(mil./$)"] / economy["population(mil.)"]
)
economy
```

```python
economy["GDP_per_capita"] = economy["GDP_per_capita"].round(2)
economy["real_consumption_per_capita"] = economy["real_consumption_per_capita"].round(2)
economy
```
