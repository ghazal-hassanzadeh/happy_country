```python
import pandas as pd
import country_converter as coco
```

```python
df1 = pd.read_csv("happiness.csv")   
df2 = pd.read_csv("economy.csv")  
```

```python
df1["country"] = df1["country"].replace({
    "D.R. of the Congo": "Democratic Republic of the Congo",
    "DR Congo": "Democratic Republic of the Congo",
    "Congo (Kinshasa)": "Democratic Republic of the Congo"
})

df2["country"] = df2["country"].replace({
    "D.R. of the Congo": "Democratic Republic of the Congo",
    "DR Congo": "Democratic Republic of the Congo",
    "Congo (Kinshasa)": "Democratic Republic of the Congo"
})

# Then convert using coco
df1["country_clean"] = coco.convert(names=df1["country"], to="name_short")
df2["country_clean"] = coco.convert(names=df2["country"], to="name_short")
```

```python
print(df1[["country", "country_clean"]].head())
print(df2[["country", "country_clean"]].head())
```

```python
merged = df1.merge(df2, on="country_clean", how="inner")
```

```python
merged.head()
```

```python
df1.to_csv("happiness_cleaned.csv", index=False)
df2.to_csv("economy_cleaned.csv", index=False)
```
